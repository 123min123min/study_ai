```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. 데이터 수집/생성 (예: 정규 분포에서 샘플링)
np.random.seed(42)
N = 100
samples = np.random.normal(loc=0, scale=1, size=N)  # 표준 정규 분포에서 1000개 샘플 생성

# 2. 경험적 CDF 계산
def empirical_cdf(x, samples):
    return np.mean(samples <= x)

# 3. 경험적 확률 계산 (예: P(-1 ≤ X ≤ 1))
def empirical_probability(a, b, samples):
    return np.mean((samples >= a) & (samples <= b))

# 4. 시각화
x = np.linspace(-3, 3, 1000)
empirical_cdf_values = [empirical_cdf(xi, samples) for xi in x]

plt.figure(figsize=(10, 6))
plt.plot(x, empirical_cdf_values, label='Empirical CDF')
plt.plot(x, stats.norm.cdf(x), label='True CDF (standard normal)', linestyle='--')
plt.legend()
plt.title("Comparison of Empirical CDF and True CDF")
plt.grid(True)
plt.show()

# 결과 출력
print(f"Empirical probability of -1 ≤ X ≤ 1: {empirical_probability(-1, 1, samples):.4f}")
# 참고: 표준 정규 분포에서 -1과 1 사이의 실제 확률은 약 0.6827
```
